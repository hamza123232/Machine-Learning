```python
# importing libraries
import matplotlib.pyplot as plt
import seaborn as sns

# loading data-set {default data-set in seaborn}

flower = sns.load_dataset("iris")
# calling flower to print
flower
```

```python
## Adding title
```

```python

# drawing a line plot
sns.lineplot(x="sepal_length",y="sepal_width", data = flower)
plt.title("Flower Graph")
plt.show()
```

```python
## Adding limits

sns.lineplot(x="sepal_length",y="sepal_width", data = flower)
plt.title("Flower Graph")
plt.xlim(4)
plt.ylim(1)
plt.show()
```

```python
ship = sns.load_dataset("titanic")
ship
```

```python
sns.barplot(x="survived",y="age" ,hue="sex", data=ship)
plt.show()
```

```python
sns.barplot(x="survived",y="age" ,hue="sex", data=ship,color="grey")
plt.show()
```

```python
sns.barplot(x="survived",y="age" ,hue="sex", data=ship,color="brown",ci= None)
plt.show()
```

```python
sns.barplot(x="survived",y="age" ,hue="sex", data=ship,color="brown",ci= None,palette='pastel')
plt.show()
```

```python
from statistics import mean, median
import numpy 

sns.barplot(x="survived",y="age" ,hue="sex", data=ship,color="brown",ci= None,palette='pastel',estimator=median)
plt.show()
```

```python
sns.barplot(x="survived",y="age" ,hue="sex", data=ship,color="brown",ci= None,palette='pastel',saturation=2)
plt.show()
```

```python
sns.barplot(x="fare",y="class" ,hue="sex", data=ship,color="brown",ci= None,palette='pastel',saturation=2)
plt.show()
```
