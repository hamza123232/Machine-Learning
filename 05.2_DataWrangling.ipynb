Data Normalization

```python
titanic.head()
```

```python
titanic1 = titanic[['age','fare']]
titanic1
```

```python
# simple feature scaling
titanic1['fare'] = titanic1['fare']/titanic1['fare'].max()
```

```python
titanic1['fare']
```

```python
# min max
titanic1['age'] = (titanic1['age']-titanic1['age'].min()/ titanic1['age'].max() - titanic1['age'].min())
```

```python
titanic1['age']
```

```python
# z-score
titanic1['age'] = (titanic1['age']-titanic1['age'].mean() / titanic1['age'].std())
```

```python
titanic1['age']
```

```python
# log transformation

titanic1['fare'] = np.log(titanic1['fare'])
```

```python
titanic1['fare']
```

Binning

```python
bins = np.linspace(min(titanic1['age']),max(titanic1['age']),15000)
age_group = ['younger','elder']
titanic1['age'] = pd.cut(titanic1['age'],bins,labels=age_group,include_lowest=True)
```
